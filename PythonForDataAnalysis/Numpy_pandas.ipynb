##4.1 Numpy ndarray : 다차원 배열 객체

ndarray라고 하는 N차원의 배열 객체로 사용할 수 있는 대규모 데이터 집합을 담을 수 있는 빠르고 유연한 자료구조

배열은 스칼라 원소 간의 연산에 사용한느 문법과 비슷한 방식을 사용해서 전체 데이터 블록에 수학적인 여산을 수행할 수 있도록 해줌

```python
import numpy as np
```

```python
#임의의 값 생성
data = np.random.randn(2,3)
data
```

```python
data * 10
```

```python
data + data
```

ndarray는 같은 종류의 데이터를 담을 수 있는 포괄적인 다차원 배열 모든 원소는 같은 자료형이어야 한다.

```python
data.shape
```

```python
data.dtype
```

###ndarray의 dtype

```python
arr1 = np.array([1,2,3], dtype=np.float64)
arr2 = np.array([1,2,3], dtype=np.int32)
```

```python
arr1.dtype
```

```python
arr2.dtype
```

```python
arr = np.array([1,2,3,4,5])
arr.dtype
```

```python
float_arr = arr.astype(np.float64)
float_arr.dtype
```
