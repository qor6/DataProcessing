####덕 타이핑

```python
# 객체가 어떤 메서드나 행동을 지원하는지 알려줌
def isiterable(obj):
  try:
    iter(obj)
    return True
  except TypeError:
    return False
```

```python
isiterable('a string')
```

```python
isiterable([1,2,3])
```

```python
isiterable(5)
```
