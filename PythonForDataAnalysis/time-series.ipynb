###11.1 날짜, 시간 자료형, 도구

```python
from datetime import datetime
```

```python
now = datetime.now()
now
```

```python
now.year, now.month, now.day
```

```python
delta = datetime(2011, 1, 7) - datetime(2008, 6, 24, 8, 15)
delta
```

```python
delta.days
```

```python
delta.seconds
```

```python
from datetime import timedelta
```

```python
start = datetime(2011, 1, 7)
start
```

```python
start + timedelta(12)
```

```python
start - 2 * timedelta(12)
```
